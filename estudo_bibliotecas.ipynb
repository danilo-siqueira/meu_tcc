```python
import folium
```

```python
# Mapa de SP
# Latitude 23°33'01'' sul e na longitude 46°38'02'' oeste
m = folium.Map(location=[-23.5291, -46.6230])
```

```python
m
```

```python
# Aumentando o Zoom
m = folium.Map(location=[-23.5291, -46.6230], zoom_start=13, min_zoom=13,max_zoom=15)
```

```python
m
```

```python
# Adicionando Marcadores
m = folium.Map(location=[-23.5499, -46.6230], zoom_start=13, min_zoom=13,max_zoom=15)

folium.Marker(
    location=[-23.588333, -46.658890],
    popup="Parque Ibirapuera",
    icon=folium.Icon(),
).add_to(m)

folium.CircleMarker(
    location=[-23.588333, -46.658890],
    radius=60,
    color="#3186cc",
    fill=True,
    fill_color="#3186cc",
).add_to(m)
```
